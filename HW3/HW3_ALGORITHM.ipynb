```python
class Points:
    
    def __init__(self, node, left, right, dim = 2):
        self.coord = node[0:dim]
        self.power = node[dim]
        
        self.left_branch = left
        self.right_branch = right
        
        self.status = 0
    
    def use(self): 
        if self.status == 0:
            self.status = 1
        else:
            self.status = self.status



#It was implemented at Wikipedia page of kd-trees.
def kdtree(nodes, depth=0, dim = 2):
    try:
        k = len(nodes[0]) 
    except IndexError as e: 
        return None
    axis = depth % dim
    nodes.sort(key=lambda x: x[axis])
    return Points(nodes[int(len(nodes)/2)],kdtree(nodes[:int(len(nodes)/2)], depth + 1),kdtree(nodes[int(len(nodes)/2) + 1:], depth + 1))


class PKDQuery:
    
    def __init__(self, rect):
        self.xl,self.yl, self.xr, self.yr = rect
    
    def is_inside_polyhedron(self, point):
        if point[0] <= self.xr:
            if point[0] >= self.xl:
                if point[1] <= self.yl:
                    if point[1] >= self.yr:
                        return 1
            else:
                return 0
        else:
            return 0
    
    def max_value(self, ax):
        
        if ax == 0:
            return self.xr
        
        if ax == 1:
            return self.yl
        
        else:
            return None
        
    def min_value(self, ax):
        
        if ax == 0:
            return self.xl
        
        if ax == 1:
            return self.yr
        
        else:
            return None
        
    
# Implemented from lecture method
def powers(tree, query, lvl = 0):
    if tree is not None:
        p = 0
        
        if query.min_value(lvl % 2) <= tree.coord[lvl % 2]:
            p = p + powers(tree.left_branch, query, lvl + 1)

            if query.is_inside_polyhedron(tree.coord):
                if tree.status == 0:
                    p = p + tree.power
                    tree.use()

        if tree.coord[lvl % 2] < query.max_value(lvl % 2):
            p = p + powers(tree.right_branch, query, lvl + 1)
            if query.is_inside_polyhedron(tree.coord):
                if tree.status == 0:
                    p = p + tree.power
                    tree.use()
        return p
    
    else: 
        return 0

def total_power(tree, query):
    p = 0
    
    for item in query:
        p = p + powers(tree, PKDQuery(item))
    return p

def answer(file):
    f = open(file, "r") 
    out = open("output.txt", "w")
    k = int(f.readline())
    mine = []
    
    for i in range(k):
        x, y, power = list(map(int, f.readline().split()))
        node = [x,y,power]
        mine.append(node)
    
    n = int(f.readline())
    
    for i in range(n):
        query = []
        
        m = int(f.readline())
        for j in range(m):
            x1, y1, x2, y2 = list(map(int, f.readline().split()))
            rect = [x1,y1,x2,y2]
            query.append(rect)
        tree = kdtree(mine)
        ans = str(total_power(tree, query))
        
        if i < n - 1:
            out.write(ans + '\n')
        
        if i == n - 1:
            out.write(ans)
    f.close()
    out.close()
    return 0

answer('input.txt')
```
